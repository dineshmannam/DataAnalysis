### Q6 Graph number of movies categorized by age restrictions before 2000 and after 2000

```python
YearGroups = tvshowRatings.Year.unique()
```

```python
YearGroups
```

```python
YearGroups.min()
```

```python
YearGroups.max()
```

```python
Before2000 = tvshowRatings[(tvshowRatings.Year <= 2000)]
```

```python
Before2000
```

```python
Before2000.count()
```

```python
After2000 = tvshowRatings[(tvshowRatings.Year > 2000)]
```

```python
After2000.count()
```

```python
MoviesBefore2000 = Before2000.groupby('Age').Title.count()
```

```python
MoviesBefore2000
```

```python
MoviesAfter2000 = After2000.groupby('Age').Title.count()
```

```python
MoviesAfter2000
```

```python
plt.plot(MoviesBefore2000, 'g--')
plt.plot(MoviesAfter2000, 'r')
plt.xlabel("Age Categories")
plt.ylabel("Number of Movies")
plt.show()
```

```python
PercentageBefore2000 = MoviesBefore2000 / sum(MoviesBefore2000)
```

```python
PercentageAfter2000 = MoviesAfter2000 / sum(MoviesAfter2000)
```

```python
plt.plot(PercentageBefore2000, 'g--')
plt.plot(PercentageAfter2000, 'r')
plt.xlabel("Age Categories")
plt.ylabel("Number of Movies")
plt.show()
```
